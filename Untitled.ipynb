```python
# the following line is only necessary if working with "ipython notebook"
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt
```

```python
STORE={}
EDs = []
Rs  = []
Energies_1 = {}
Energies_2 = {}
Energies_3 = {}
Energies_4 = {}
Energies_5 = {}
Energies_6 = {}
Energies_7 = {}

with open("my_chi2.txt","r") as f:
    for line in f:
        l = line[:-1]
        if "ED=" in l:
            ED, RKAP = (l.split(",")[0]).split("=")[1] , (l.split(",")[1]).split("=")[1]
            if ED not in STORE.keys(): STORE[ED]={}
            if RKAP not in STORE[ED].keys(): STORE[ED][RKAP]=0.0
                
            if ED not in Energies_1.keys(): Energies_1[ED]={}
            if RKAP not in Energies_1[ED].keys():  Energies_1[ED][RKAP]=0.0
            if ED not in Energies_2.keys(): Energies_2[ED]={}
            if RKAP not in Energies_2[ED].keys():  Energies_2[ED][RKAP]=0.0
            if ED not in Energies_3.keys(): Energies_3[ED]={}
            if RKAP not in Energies_3[ED].keys():  Energies_3[ED][RKAP]=0.0
            if ED not in Energies_4.keys(): Energies_4[ED]={}
            if RKAP not in Energies_4[ED].keys():  Energies_4[ED][RKAP]=0.0
            if ED not in Energies_5.keys(): Energies_5[ED]={}
            if RKAP not in Energies_5[ED].keys():  Energies_5[ED][RKAP]=0.0
            if ED not in Energies_6.keys(): Energies_6[ED]={}
            if RKAP not in Energies_6[ED].keys():  Energies_6[ED][RKAP]=0.0
            if ED not in Energies_7.keys(): Energies_7[ED]={}
            if RKAP not in Energies_7[ED].keys():  Energies_7[ED][RKAP]=0.0
            
                
            if float(ED) not in EDs: EDs.append(float(ED))
            if float(RKAP) not in Rs: Rs.append(float(RKAP))
                
        elif "Chi2" in l:
            Chi2 = l.split("=")[1].replace(" ","")
            STORE[ED][RKAP]= Chi2
            
        if "Exp = 0.123" in l:
            lsp =l.replace("Exp = 0.123","")
            Energies_1[ED][RKAP] = float(lsp.split("=")[1].replace(" ",""))
        elif "Exp = 0.385" in l:
            lsp =l.replace("Exp = 0.385","")
            Energies_2[ED][RKAP] = float(lsp.split("=")[1].replace(" ",""))
        elif "Exp = 0.76" in l:
            lsp =l.replace("Exp = 0.76","")
            Energies_3[ED][RKAP] = float(lsp.split("=")[1].replace(" ",""))
        elif "Exp = 1.223" in l:
            lsp =l.replace("Exp = 1.223","")
            Energies_4[ED][RKAP] = float(lsp.split("=")[1].replace(" ",""))
        elif "Exp = 1.753" in l:
            lsp =l.replace("Exp = 1.753","")
            Energies_5[ED][RKAP] = float(lsp.split("=")[1].replace(" ",""))
        elif "Exp = 2.33" in l:
            lsp =l.replace("Exp = 2.33","")
            Energies_6[ED][RKAP] = float(lsp.split("=")[1].replace(" ",""))
        elif "Exp = 2.899" in l:
            lsp =l.replace("Exp = 2.899","")
            Energies_7[ED][RKAP] = float(lsp.split("=")[1].replace(" ",""))
print(EDs)
print(Rs)
X,Y = np.meshgrid( EDs , Rs)
Z = []
E_1=[]
E_2=[]
E_3=[]
E_4=[]
E_5=[]
E_6=[]
E_7=[]

#for x_i in EDs:
#    Z.append( [ float(STORE[str(x_i)][str(r)]) if r!=0.0 else float(STORE[str(x_i)][str("0")])  for r in Rs ]  )
for y_i in Rs:
    if y_i==0.0: y_i="0"
    Z.append( [ float(STORE[str(x)][str(y_i)]) for x in EDs] )
    E_1.append( [float(Energies_1[str(x)][str(y_i)]) for x in EDs] )
    E_2.append( [float(Energies_2[str(x)][str(y_i)]) for x in EDs] )
    E_3.append( [float(Energies_3[str(x)][str(y_i)]) for x in EDs] )
    E_4.append( [float(Energies_4[str(x)][str(y_i)]) for x in EDs] )
    E_5.append( [float(Energies_5[str(x)][str(y_i)]) for x in EDs] )
    E_6.append( [float(Energies_6[str(x)][str(y_i)]) for x in EDs] )
    E_7.append( [float(Energies_7[str(x)][str(y_i)]) for x in EDs] )
    
```

```python
fig, ax = plt.subplots()

ax.scatter(X, Y, color="green")
ax.set_title('Regular Grid, created by Meshgrid')
ax.set_xlabel('Ed')
ax.set_ylabel('Rk')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 

levels = np.arange(0.0,20,1)

point =[0.5,-0.08]

cp = ax.contour(X, Y, Z, levels=levels)
ax.clabel(cp, inline=True, 
          fontsize=10)
ax.set_title('Chi2Value')
ax.set_xlabel('Ed')
ax.set_ylabel('Rk')

plt.scatter( [0.5], [-0.08], color="r", marker="+")

plt.show()
```

```python
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)

ax.set_title('Contour Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```
